```python
bins = [0,5,10,15,20,25,30]
labels = ["0-5","5-10","10-15","15-20","20-25","25-30"]

df["pts_group"] = pd.cut(df["pts"], bins=bins, labels=labels)

freq = df["pts_group"].value_counts()
freq

(freq / len(df)) * 100
```

This groups players into scoring ranges.

```python
# Select needed columns
cols = ["pts", "reb", "ast", "net_rating", "usg_pct", "ts_pct"]

# Correlation matrix
corr = df[cols].corr()
corr
```

```python
season_avg = df.groupby("season")[["pts","reb","ast"]].mean()

print(season_avg)
```

```python
yoy_change = season_avg.diff()

print("Year-over-Year Change:")
print(yoy_change)
```

```python
rolling_mean = season_avg.rolling(3).mean()

print("3-Season Rolling Mean:")
print(rolling_mean)
```

```python
df["usage_group"] = pd.qcut(df["usg_pct"], 4, labels=["Q1","Q2","Q3","Q4"])
```

```python
group_stats = df.groupby("usage_group")["pts"].agg(["mean","std","skew"])
print(group_stats)
```

```python
five_num = df.groupby("usage_group")["pts"].quantile([0,0.25,0.5,0.75,1]).unstack()
five_num.columns = ["min","Q1","median","Q3","max"]

print(five_num)
```
